### Filtering our data

```python
# Find specific values
Sales_CSV[Sales_CSV['Payment Type'] == 'Cash']
# Using the logical operator "&"
Sales_CSV[(Sales_CSV['Rating'] >= 3) & (Sales_CSV['Rating'] <= 7)]
# Using the logical operator "|"
Sales_CSV[(Sales_CSV['Branch'] == 'A') | (Sales_CSV['Branch'] == 'C')]
```

#### Know the number of records of the filtered data.

```python
print(len(Sales_CSV[Sales_CSV['Payment Type'] == 'Cash']))
print(len(Sales_CSV[(Sales_CSV['Rating'] >= 3) & (Sales_CSV['Rating'] <= 7)]))
print(len(Sales_CSV[(Sales_CSV['Branch'] == 'A') | (Sales_CSV['Branch'] == 'C')]))
```

#### Know the number of records of the filtered data (using the method "between")

### Using functions
Create a function that divide the value of the data by 2

```python
def marks(x):
    return x/2
```

Using our created function and appling to our data

```python
Sales_CSV['Unit Price'].apply(marks)
```

Also, we able to add a new column applying some function created by us.

```python
Sales_CSV['Half Unit Price'] = Sales_CSV['Unit Price'].apply(marks)
Sales_CSV
```

Or, we can use <code>"lambda"</code> for apply some function to each element of a specific column and to get a new values.

```python
Sales_CSV['Unit Price'].apply(lambda x: x/2)
```

Or, we can use <code>"apply"</code> to apply built-in function to our dataset.

```python
Sales_CSV['City'].apply(len)
```

### Assign general new values to represent repeated values in a same column.

```python
Sales_CSV['Gender'].map({'Male':1, 'Female':0})
```

Add this new representation of values as a new column in our dataset.

```python
Sales_CSV['Male_Female'] = Sales_CSV['Gender'].map({'Male':1, 'Female':0})
Sales_CSV
```

### Delete column(s)
With the <code>"inplace"</code> argument, not necessary assign again to the DataFrame main variable.

```python
Sales_CSV.drop(['Half Unit Price', 'Male_Female'], axis=1, inplace=True)
Sales_CSV
```

By default: If not use the "inplace" argument, then assign againt to the DataFrame variable.

```python
Sales_CSV = Sales_CSV.drop(['Half Unit Price', 'Male_Female'], axis=1, inplace=False)
Sales_CSV
```
